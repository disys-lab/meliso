```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator

plt.rcParams['font.family'] = 'serif'
plt.rcParams['figure.dpi'] = 1000
plt.rcParams['font.size'] = 14
plt.rcParams['legend.fontsize'] = 14
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['xtick.major.size'] = 5.0
plt.rcParams['xtick.minor.size'] = 3.0
plt.rcParams['ytick.major.size'] = 5.0
plt.rcParams['ytick.minor.size'] = 3.0
plt.rcParams['axes.linewidth'] = 0.8
plt.rcParams['legend.handlelength'] = 2.0
```

```python
data = {
    "Material": ["Ag-aSi", "Ag-aSi", "Ag-aSi", "Ag-aSi", "Ag-aSi", 
                 "AlOx-HfO2", "AlOx-HfO2", "AlOx-HfO2", "AlOx-HfO2", "AlOx-HfO2",
                 "EpiRAM", "EpiRAM", "EpiRAM", "EpiRAM", "EpiRAM",
                 "TaOx-HfOx", "TaOx-HfOx", "TaOx-HfOx", "TaOx-HfOx", "TaOx-HfOx"],
    "No. iteration": [1, 3, 5, 10, 100, 1, 3, 5, 10, 100, 1, 3, 5, 10, 100, 1, 3, 5, 10, 100],
    "L-2 Norm Error (Mean)": [24.27221, 10.5966, 4.5625, 0.6102, 0.563618,
                              51.3709, 2.1067, 2.8065, 1.8870, 1.6796,
                              1.9780, 3.1863, 3.1662, 3.0770, 3.1295,
                              34.45485, 1.9522, 2.004344, 2.004161, 2.012602],
    "L-infty Norm Error (Mean)": [3.3054, 1.5956, 0.8019, 0.1882, 0.197282,
                                  6.6886, 0.4079, 0.6746, 0.3762, 0.4198,
                                  0.3775, 0.6557, 0.6523, 0.6523, 0.6523,
                                  4.581318, 0.3386, 0.343384, 2.004161, 0.342148]
}

df = pd.DataFrame(data)
materials = df['Material'].unique()

for material in materials:
    material_data = df[df['Material'] == material]
    
    plt.figure(figsize=(10, 6))
    
    plt.plot(material_data['No. iteration'], material_data['L-2 Norm Error (Mean)'], marker='o', label='L-2 Norm Error (Mean)')
    plt.plot(material_data['No. iteration'], material_data['L-infty Norm Error (Mean)'], marker='o', label='L-infty Norm Error (Mean)')
    
    plt.title(f'Errors vs Number of Iterations for {material}')
    plt.xlabel('Number of Iterations')
    plt.ylabel('Error (Mean)')
    plt.legend()
    plt.grid(True)
    
    plt.show()

```
